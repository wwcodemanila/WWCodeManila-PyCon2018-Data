## What is Jupyter Notebook?

Jupyter Notebook is a browser-based tool that can contain texts, codes, graphics, equations, rich media, and others. It is an open-source interactive computing tool that is being developed and maintained by [Project Jupyter](http://jupyter.org/).

It is composed of:
1. **Kernel** - which is the component that runs and interprets the code. Jupyter supports a variety of kernels including Fortran, PySpark, Python, Julia, Haskell, R, Ruby, Javascript, and [more](https://github.com/jupyter/jupyter/wiki/Jupyter-kernels).
2. **Dashboard** - the graphical interface where you can add, delete, edit, and manage both Python notebooks and kernels. Try the tour feature by going over to `Help` in the menu above, and select `User interface tour`

## Keyboard shortcuts

There are two types of modes when working on Jupyter cells:

1. Edit mode (`Enter`) - the cell is highlighted in green and you can write within the cell
2. Command mode (`Esc`) - the cell is highlighted in blue and you cannot write inside the cell. However, you can use keyboard shortcuts

**Some useful keyboard shortcuts**

`Shift + Enter` - run cell, and jump to the cell below  
`Ctrl + Enter` - run cell in place  
`Alt + Enter` - run cell and insert new cell blow  
`A` - insert cell above  
`B` - insert cell below  
`C` - copy cell  
`V` - paste cell  
`DD` - delete selected cell  
`Z` - undo delete
`Y` - change cell to Code mode  
`M` - change cell to Markdown mode  
`Ctrl + Shift + P` or `Ctrl + Shift + F` - access the command pallete
